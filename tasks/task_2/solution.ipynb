```python
import os
import hashlib
```

```python
def calculate_sha3_256(file_path='task2_data/file_00.data'):
    sha3_256_hash = hashlib.sha3_256()
    with open(file_path, 'rb') as f:
        data = f.read()
        sha3_256_hash.update(data)

    return sha3_256_hash.hexdigest()

def process_directory(folder_path):

    hashes = []

    for filename in sorted(os.listdir(folder_path)):
        file_path = os.path.join(folder_path, filename)

        if os.path.isfile(file_path):
            try:
                hash_value = calculate_sha3_256(file_path)
                sorted_hash = 1
                for char in hash_value:
                    sorted_hash *= int(char, 16) + 1
                hashes.append(str(sorted_hash))
        
            except IOError as e:
                print(f"Error processing file {file_path}: {e}")

    return hashes

hashes = process_directory('task2_data')
```

```python
import hashlib
import os

def sha3_file(path):
    h = hashlib.sha3_256()
    with open(path, "rb") as f:
        h.update(f.read())
    return h.hexdigest()

def sort_key(hex_hash):
    product = 1
    for ch in hex_hash:
        product *= int(ch, 16) + 1
    return product

def process(folder, email):
    pairs = []

    for filename in sorted(os.listdir(folder)):
        p = os.path.join(folder, filename)
        if os.path.isfile(p):
            h = sha3_file(p)
            key = sort_key(h)
            pairs.append((h, key))

    pairs.sort(key=lambda x: x[1])

    joined = "".join(h for (h, _) in pairs)

    final_input = joined + email.lower()

    final_hash = hashlib.sha3_256(final_input.encode()).hexdigest()
    return final_hash, final_input

final_hash, final_input = process("task2_data", "roostamovic@yahoo.com")
print(final_hash, final_input)
```
